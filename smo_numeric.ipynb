```python
time_init           = 1
time_final          = 10**8

n_particles         = 2000

diffusion_exponent  = -1.0
area_exponent       =  0.0

coag_const          = 150
frag_const          = 0.0
```

```python
result_dir = os.path.join("diff_%.2f"%diffusion_exponent, "area_%.2f"%area_exponent, 
                          "coag_"+str(coag_const), "frag_"+str(frag_const))
print(result_dir)
```

<h1> Mean Cluster Approach </h1>

<h2> Parameters </h2>

```python
m_coag_const        = coag_const / 1500
m_frag_const        = frag_const

m_max               = n_particles

m0                  = 1.0
m_dt_0              = 10**-6

dimensions          = 2
nu                  = diffusion_exponent

exponent_frag       = 0.0 #3.7

time_snap           = 10
time_change_dt      = 1000
```

```python
exponent_coag       = (1.0-(2.0/dimensions)+nu)
```

```python
mca_result_dir = os.path.join(result_dir, "mca","")

try:
    os.makedirs(mca_result_dir)
except:
    pass
```

<h2> Equation </h2>

$$ \frac{dm}{dt} = C m^{(1.0-2.0/d+\nu)} \bigg( 1.0 - \frac{m}{m_{max}} \bigg) - F (m^{expo\_{frag}})$$

<h2> Numerical Integration </h2>

```python
time  = time_init
count = 0
m     = m0
m_old = m0

m_dt  = m_dt_0

m_time_list = []
m_mca_list  = []

output_file = open(mca_result_dir+r"mca_result.dat","w")

while time < time_final:
    m = m_old + m_dt * (m_coag_const*((m_old**exponent_coag)*(1.0-m_old/m_max))  )#- m_frag_const*(mt.e**m_old))

    if count % time_snap == 0:
        output_string = str(time)+" "+str(m)
        output_file.write(output_string+"\n")
        print(output_string+"                                          ", end = "\r")
        m_time_list.append(time)
        m_mca_list.append(m)
        if abs(n_particles - m) < 0.01:
            break
    
    m_old = m
    
    
    time += m_dt
    if count % time_change_dt == 0:
        m_dt *= 10 
        
    
    count += 1

output_file.close() 

m_real_time_final = time
```

```python
try:
    output_file.close()   
except:
    print("file already closed")
```

<h2> Functions </h2>

```python
# coagulation kernel                                                                                                                          

def area_term(mass_i, mass_j, exponent):
    area_i = mass_i**exponent
    area_j = mass_j**exponent
    return ((area_i + area_j) / 2.0)

def diffusion_term(mass_i, mass_j, exponent):
    diff_i = mass_i ** exponent
    diff_j = mass_j ** exponent
    return ((diff_i + diff_j) / 2.0)


def size_difference(mass_i, mass_j, n_particles):
    exponent = 10
    return (1.0 - (mt.sqrt((mass_i - mass_j)**2)/n_particles))**exponent


def coagulation_kernel(mass_i, mass_j, diffusion_exponent, area_exponent, n_particles):
    area      = area_term(mass_i, mass_j, area_exponent)
    diffusion = diffusion_term(mass_i, mass_j, diffusion_exponent)
    return  area * diffusion

# fragmentation kernel

def fragmentation_kernel(mass_i, mass_j, exponent):
    mt.e ** (- ((mass_i**0.5 + mass_j**0.5) - (mass_i + mass_j)**0.5) / temperature  )
    
# cluster count                                                                                                                               

def number_of_clusters(n_k, n_particles):
    num_cluster = 0
    for idx_k in range(n_particles + 1):
        num_cluster += n_k[idx_k]
    return num_cluster

# cluster_medio                                                                                                                               
def cluster_medio(cluster_list = [], n_k = []):
    if len(cluster_list) > 0:
        return sum(cluster_list)/len(cluster_list)
    if len(n_k) > 0:
        n_particles = sum([k * n_k[k] for k in range(0, len(n_k))])
    return n_particles/sum(n_k)

# cluster distribution                                                                                                                        
def mass_distribution(cluster_list, n_particles):
    n_k = [0]
    for k in range(1, n_particles+1):
        n_k.append(0)
        for cluster in cluster_list:
            if cluster == k:
                n_k[k] += 1
    return n_k

# non-zero distribution                                                                                                                       
def non_zero_dist(n_k):
    n_k_non_zero = {}
    for idx_k in range(1, len(n_k)):
        if n_k[idx_k] != 0:
            n_k_non_zero[idx_k] = n_k[idx_k]
    return n_k_non_zero
```

<h2> Calculating Coagulation and Fragmentation Kernels Matrix </h2>

```python
matriz_coagulation_kernel = np.zeros((n_particles + 1, n_particles + 1))
matriz_fragmentation_kernel = np.zeros((n_particles + 1, n_particles + 1))

for mass_i in range(1, n_particles + 1):
    for mass_j in range(1, n_particles + 1):
        matriz_coagulation_kernel[mass_i,mass_j] = coagulation_kernel(mass_i, mass_j, diffusion_exponent, area_exponent, n_particles)
        matriz_fragmentation_kernel[mass_i,mass_j] = fragmentation_kernel(mass_i, mass_j, 1.0)
        if USE_DISTANCE:
            matriz_coagulation_kernel[mass_i,mass_j] *= size_difference(mass_i, mass_j, n_particles)
```

$$ \frac{dn_k}{dt} =  \frac{1}{2} \displaystyle\sum_{i=1}^{k-1} 
\left( {\cal{C}}_{i,k-i}n_in_{k-i} - {\cal {F}}_{i,k-i}n_k \right)
- \displaystyle\sum_{j=1}^\infty 
\left( {\cal{C}}_{j,k}n_jn_{k} - {\cal {F}}_{j,k}n_{j+k} \right)   $$

```python
# Discrete smoluchowski simulation

if DISCRETE:
    n_k          = [ n_particles if idx == 1 else 0  for idx in range(0, n_particles + 1)]
    n_k_non_zero = non_zero_dist(n_k)
    
    time_stable   = 0
    count_stable  = 0
    k_mean        = cluster_medio(n_k = n_k)

    delta_image   = 0.2
    next_image    = 0

    s_time_list     = []
    s_smo_list      = []

    time = time_init
    output_file = open(smo_disc_result_dir+r"smo_result.dat","w")
    while time < time_final:
        # size by size                                                                                                                            
        for mass_i in n_k_non_zero:
            for mass_j in n_k_non_zero:
                if n_k[mass_i] > 0 and n_k[mass_j] > 0:
                    merge_score = random.random()

                    coagulation_probability =  multiplier * (n_k[mass_i]/n_particles) * (n_k[mass_j]/n_particles)
                    coagulation_probability *= matriz_coagulation_kernel[mass_i,mass_j] 



                    if merge_only_similar:
                        # impedindo agregados diferentes demais de agregar                                                                        
                        if (mass_i / mass_j < cluster_fusion_frac or mass_j / mass_i < cluster_fusion_frac) and (mass_i > small_cluster_size and mass_j > small_cluster_size):
                            #print()                                                                                                              
                            coagulation_probability = 0.0

                    if coagulation_probability > 1:
                        number_merges = int(coagulation_probability)
                        if mass_i == mass_j:
                            if (2 * n_k[mass_i]) < number_merges:
                                number_merges = int(n_k[mass_i] / 2)
                        else:
                            if n_k[mass_i] < number_merges:
                                number_merges = n_k[mass_i]
                            if n_k[mass_j] < number_merges:
                                number_merges = n_k[mass_j]
                    elif merge_score < coagulation_probability:
                        number_merges         = 1
                        if mass_i == mass_j:
                            if n_k[mass_i] < 2:
                                number_merges = 0
                    else:
                        number_merges = 0
                    if number_merges > 0:
                        n_k[mass_i]          -= number_merges
                        n_k[mass_j]          -= number_merges

                        if n_k[mass_i] < 0:
                            print("# ", mass_i, n_k[mass_i], number_merges)
                        if n_k[mass_j] < 0:
                            print("# ", mass_j, n_k[mass_j], number_merges)

                        try:
                            if mass_i + mass_j <= n_particles:
                                n_k[mass_i + mass_j] += number_merges
                            else:
                                print(" # ", mass_i, mass_j, n_k[mass_i], number_merges)
                        except:
                            print("# ", mass_i, mass_j, mass_i + mass_j)
                            n_k[mass_i + mass_j] += number_merges

        n_k_non_zero = non_zero_dist(n_k)

        if time >= next_image:
            k_mean        = cluster_medio(n_k = n_k)
            output_string = str(time)+"\t"+"%.2f "%(k_mean)
            output_file.write(output_string+"\n")
            print(output_string, end = "\r")

            s_time_list.append(time)
            s_smo_list.append(k_mean)

            next_image += delta_image*10**(mt.log10(time))
            if k_mean > n_particles - 1 and time_stable == 0:
                time_stable = time
            elif k_mean > n_particles - 1 and time_stable != 0:
                count_stable += 1
        if count_stable > 5:
            break
        time += 1                
    output_file.close()   
```
